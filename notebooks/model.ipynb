```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

```python
# Loading dataset
df = pd.read_csv(r'C:\Users\Hp\Desktop\work\capstone-project-RupiaMorris\data\preprocessed_climate_data.csv')
df.head()
```

```python
df = df.drop('Date', axis=1)
```

```python
# Define Features and Target Variable
x = df.drop(columns = ['Total_Rainfall_mm'])
y = df['Total_Rainfall_mm']

# Train-test split (80% train and 20% test)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 42)
```

```python
# Linear Regression Model as model_1
model_1 = LinearRegression()
model_1.fit(x_train, y_train)
```

```python
# Random Forest
model_2= RandomForestRegressor(random_state=42)
model_2.fit(x_train, y_train)
```

```python
# Linear Regression Predictions
lr_preds = model_1.predict(x_test)

# Random Forest Predictions
rf_preds = model_2.predict(x_test)

# Evaluation function
def evaluate_model(y_true, y_pred, model_name):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    print(f"{model_name} - MAE: {mae:.2f}, RMSE: {rmse:.2f}")

# Evaluate both
evaluate_model(y_test, lr_preds, "Linear Regression")
evaluate_model(y_test, rf_preds, "Random Forest")
```

```python
importances = model_2.feature_importances_
features = x_train.columns

plt.figure(figsize=(8, 4))
plt.barh(features, importances)
plt.xlabel('Feature Importance')
plt.title('Random Forest - Feature Importance')
plt.show()
```
