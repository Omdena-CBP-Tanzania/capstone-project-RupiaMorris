```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import os
```

```python
# Loading the datas
df = pd.read_csv(r'C:\Users\Hp\Desktop\work\capstone-project-RupiaMorris\data\tanzania_climate_data.csv')
df.head()
```

```python
# Checking Missing Values
df.isnull().sum()
```

```python
# data summary
df.info()
```

```python
# Setting date as index of our dataset
df['Date'] = pd.to_datetime(df[['Year', 'Month']].assign(DAY=1))
df.set_index('Date', inplace= True)

# Daily temperature (Temperature Ramge)
df['Temp_Range_C'] = df['Max_Temperature_C'] - df['Min_Temperature_C']
```

```python
# Adding Tanzania specific seasons
def get_tanzania_season(month):
    if month in [3, 4, 5]:
        return 'Long_Rains'
    elif month in [6,7,8,9, 10]:
        return "Dry_Season"
    elif month in [11, 12]:
        return "Short_Rains"
    else: # January and February
        return 'Hot_Dry'
    
df['Season'] = df['Month'].apply(get_tanzania_season)
```

```python
df.head()
```

# Encoding Categorical Variable

```python
# Initialize the encoder
encoder = LabelEncoder()

#Applying encoder to Season Column
df['Season'] = encoder.fit_transform(df['Season'])
```
